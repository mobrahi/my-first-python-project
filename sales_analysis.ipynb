```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sample sales data
data = {
    'Customer': ['TechCorp', 'DataSystems', 'CloudNine', 'TechCorp', 'BigData Co'],
    'Deal_Size': [50000, 25000, 120000, 50000, 75000],
    'Status': ['Closed', 'Pending', 'Closed', 'Closed', 'Pending']
}
# prepare list as DataFrame
df = pd.DataFrame(data)
```

```python
# Perform "Account Manager" Analysis - total revenue from closed deals
closed_deals_total = df[df['Status'] == 'Closed']['Deal_Size'].sum()

# Clean customer list by removing duplicates
clean_df = df.drop_duplicates()
```

```python
print(f"--- Sales Report ---")
print(f"Total Revenue from Closed Deals: {closed_deals_total}")
print(f"\nCleaned Customer List:\n{clean_df}")
```

```python
# Group the data by Customer for the chart
summary = df.groupby('Customer')['Deal_Size'].sum().sort_values(ascending=False)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
summary.plot(kind='bar', color='skyblue')
plt.title('Total Deal Size by Customer')
plt.xlabel('Customer')
plt.ylabel('Total Deal Size ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Save the chart as an image file

plt.tight_layout()
plt.savefig('sales_chart.png')

print("Success! Your chart has been saved as 'sales_chart.png'")
```
